```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1.1, 0.1)
print(x)
plt.figure() # 第一环节 创建画布
plt.plot(x,x**2) # 第二环节，绘制图形
plt.plot(x,x**4)
plt.show() # 第三环节，显示图形
```

```python
x = np.arange(0, 1.1, 0.1)
print(x)
plt.figure() # 第一环节 创建画布
plt.plot(x,x**2) # 第二环节，绘制图形
plt.plot(x,x**4)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title('lines')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y=x^2','y=x^4']) # 注意顺序 同时要用列表形式！

plt.savefig('tmp/examplt.png')
plt.show() # 第三环节，显示图形
```
